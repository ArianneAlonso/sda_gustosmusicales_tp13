```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Datos simples: potencia (horsepower) y consumo (mpg)
horsepower = np.array([130, 165, 150, 140, 198, 220, 215, 225, 190, 170], dtype=float)
mpg = np.array([18, 15, 18, 16, 15, 14, 14, 14, 15, 15], dtype=float)
```

```python
horsepower_norm = (horsepower - horsepower.min()) / (horsepower.max() - horsepower.min())
mpg_norm = (mpg - mpg.min()) / (mpg.max() - mpg.min())
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),  # Capa de entrada explícita
    tf.keras.layers.Dense(10, activation='sigmoid'),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
```

```python
history = model.fit(horsepower_norm, mpg_norm, epochs=50, batch_size=4, verbose=1)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

```python
xs = np.linspace(0, 1, 100)
ys = model.predict(xs)

xs_orig = xs * (horsepower.max() - horsepower.min()) + horsepower.min()
ys_orig = ys.flatten() * (mpg.max() - mpg.min()) + mpg.min()

plt.scatter(horsepower, mpg, label='Datos originales')
plt.plot(xs_orig, ys_orig, color='red', label='Predicción')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.legend()
plt.show()
```
